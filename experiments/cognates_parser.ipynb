```python
import urllib.request
import urllib.error
import time
import re
import pickle
import lxml.html as html
```

```python
def get_page(url):
    try:
        time.sleep(1)
        proxies = {'http': '183.111.169.201:8080'}
        return urllib.request.urlopen(url, timeout=5).read()
    except urllib.error.HTTPError as e:
        print(e)
        return None
    except BaseException as e:
        print(e)
        return None
```

```python
letters = ['','b','v','g','d','ye','yo','zh','z','i','yi','k','l','m','n','o','p','r','s','t','u','f','h','c','ch','sh','sc','ii','e','yu','ya']
```

```python
result = []

for letter in letters:
    url = 'http://cognates.ru/roots/' + letter
    print(url)
    page = get_page(url)
    if page is not None:
        content = html.fromstring(page)
        #words = content.xpath("//a[contains(@href,'pn')]/text()")[1:]
        #print(words)
        links = content.xpath("//a[contains(@href,'pn')]/@href")
        result.append(links)
        print(links)
```

```python
roots_pages = result.append(list(map(lambda s: '/roots/' + s, letters)))
```

```python
roots_pages = result
```

```python
#with open('roots_pages', 'wb') as f:
#     pickle.dump(roots_pages, f)
```

```python
roots_links = []

for sublist in roots_pages:
    for item in sublist:
        url = 'http://cognates.ru' + item
        page = get_page(url)
        if page is not None:
            content = html.fromstring(page)
            print(url)
            links = content.xpath("//a[contains(@href,'root')]/@href")[1:]
            print(links)
            roots_links += links
    
#    if i % 100 == 0:
#        pickle.dump(result, open("cognates" + str(i), "wb"))
```

```python
roots_links_filtered = set(filter(lambda item: re.match(".*root[0-9]+", item), roots_links))
```

```python
with open('roots_links_filtered', 'rb') as f:
    roots_links_filtered = pickle.load(f)

result = []

for i, item in enumerate(roots_links_filtered):
    url = 'http://cognates.ru' + item
    page = get_page(url)
    if page is not None:
        content = html.fromstring(page)
        print(url)
        words = content.xpath("//a[contains(@href,'words')]/text()")[1:]
        if len(words) < 2:
            continue
        print(words)
        result.append(words)
    
    if i % 100 == 0:
        pickle.dump(result, open("cognates" + str(i), "wb"))
```
